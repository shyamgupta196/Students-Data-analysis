# 3D visualisations for data science and machine learning 

## Example 1 - Matplotlib

## importing important Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import plotly.express as px
from google.colab import files
```

```python
upload = files.upload()
```

```python
df= pd.read_csv("Iris.csv")
df.head()
```

```python
df.columns
```

## data visualisation

```python
plt.figure(figsize=(10,8))
ax = plt.axes(projection = '3d')
fg = ax.scatter3D(df['SepalWidthCm'], df['SepalLengthCm'], df['PetalLengthCm'],s=df['PetalWidthCm']*50,c=df['PetalWidthCm'])

ax.set_xlabel('SepalWidthCm')
ax.set_ylabel('SepalLengthCm')
ax.set_zlabel('PetalLengthCm')

plt.colorbar(fg)
```
